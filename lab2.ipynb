```python
import re
import requests
from dataclasses import dataclass
```

```python
link = "https://en.wikipedia.org/wiki/Lord_of_War"

page = requests.get(link).text
page
```

```python
name_of_film = re.findall(r'wiki.*/(.*)/?', link)[0]

name_of_film
```

```python
country_of_film = re.findall(r'Country<\/th><td class="infobox-data">([A-Za-z\s]+)', page)[0]

country_of_film
```

```python
release_date = re.findall(r'Release date.*>\n.*(\d{4}-\d{2}-\d{2})', page)[0]

release_date
```

```python
director = re.findall(r'Directed by.*title="(.*?)"', page)[0]

director
```

```python
box_office = re.findall(r'Box office</th>.*">(.*?)<sup', page)[0]
box_office
```

```python
@dataclass
class Film:
    link: str
    name: str 
    country: str
    release: str
    director: str
    box_office: str

    def get_year(self):
        return self.release.split("-")[0]

film1 = Film(link, name_of_film, country_of_film, release_date, director, box_office)
```
